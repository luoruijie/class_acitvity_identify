```python
import os
import pandas as pd
import json
import string

# 定义提取 "key_text" 的函数
def extract_key_text(json_str):
    try:
        json_obj = json.loads(json_str)
        return json_obj.get("key_text", "")
    except json.JSONDecodeError:
        return ""

# 检查文本是否只剩下标点符号的函数
def is_only_punctuation(text):
    # 去除文本中的所有标点符号和空白字符
    text_no_punctuation = text.translate(str.maketrans('', '', string.punctuation)).strip()
    return len(text_no_punctuation) == 0

# 定义文件目录
data_dir = "data\\history"

# 获取目录下所有文件名
data_dir_list = os.listdir(data_dir)
list_concat =[]
# 循环读取每一个文件
for file_name in data_dir_list:
    if file_name.endswith('.xlsx') and not file_name.startswith('~$'):  # 忽略临时文件
        file_path = os.path.join(data_dir, file_name)
        
        # 读取Excel文件
        df = pd.read_excel(file_path)
        
        # 提取“key_text”列
        df['key_text'] = df['gpt4o_output_json'].apply(extract_key_text)
        
        # 添加 "non_key_text" 列
        def fill_non_key_text(row):
            if pd.isna(row['key_text']) or row['key_text'] == "":
                return row['text']
            else:
                # 计算 "text" 减去 "key_text" 后的结果
                remaining_text = row['text'].replace(row['key_text'], "", 1).strip()
                
                # 如果剩余的文本只有标点符号，返回空值；否则返回剩余文本
                if is_only_punctuation(remaining_text):
                    return ""
                else:
                    return remaining_text

        df['non_key_text'] = df.apply(fill_non_key_text, axis=1)
        
        # 将结果保存回原文件中
        df.to_excel(file_path, index=False)
        list_concat.append(df)

        print(f"已处理并保存文件: {file_name}")
```

```python
 df_all = pd.concat(list_concat)
```
