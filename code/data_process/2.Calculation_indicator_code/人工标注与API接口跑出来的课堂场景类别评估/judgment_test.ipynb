```python
# 定义一个函数来比较两个label值
def compare_labels(label1, label2):
    labels1 = label1.strip().replace(' ', '').split('、')
    labels2 = label2.strip().replace(' ', '').split('、')
    
    if len(labels1) == 1 and len(labels2) == 1:
        return 1 if labels1[0] == labels2[0] else 0
    else:
        set1 = set(labels1)
        set2 = set(labels2)
        return 1 if set1 & set2 else 0
```

```python
for i in range(len(df)):
    df.loc[i,"match"] = compare_labels(df.loc[i,"label"],df.loc[i,"模型预测Lable"])

   
```

```python
import pandas as pd
from sklearn.metrics import precision_score, recall_score, accuracy_score

# 读取用户上传的Excel文件
file_path = '高希娜_预测结果_vllm (4).xlsx'
df = pd.read_excel(file_path)

# 用 "NA" 填充缺失值
df_filled = df.fillna('NA')

# 定义类别标签
categories = ["个体发言", "个体展示", "独立练习", "学生齐写", "小组学习", "学生听写", "学生齐读", "多人展示", "集体未知", "个体未知"]

# 方式二：忽略带有“或”的标签
df_no_or = df_filled[~df_filled['label'].str.contains('或')]

# 初始化匹配计数和总计数
matches_no_or = {category: 0 for category in categories}
total_counts = {category: 0 for category in categories}

# 计算不包含“或”标签的匹配计数和总计数
if not df_no_or.empty:
    labels_no_or = df_no_or['label']
    predictions_no_or = df_no_or['模型预测Lable']

    for index, row in df_no_or.iterrows():
        label = row['label']
        prediction = row['模型预测Lable']
        if label in categories:
            total_counts[label] += 1
            if label == prediction:
                matches_no_or[label] += 1

# 处理带有“或”的标签
df_or = df_filled[df_filled['label'].str.contains('或')]

# 初始化带有“或”标签的匹配计数
matches_or = {category: 0 for category in categories}

# 计算带有“或”标签的匹配计数
if not df_or.empty:
    for index, row in df_or.iterrows():
        label = row['label']
        prediction = row['模型预测Lable']
        possible_labels = label.split('或')
        if prediction in possible_labels:
            matches_or[prediction] += 1
        for possible_label in possible_labels:
            if possible_label in categories:
                total_counts[possible_label] += 1

# 合并匹配计数
total_matches = {category: matches_no_or.get(category, 0) + matches_or.get(category, 0) for category in categories}

# 计算每个类别的精确度、召回率
precision_total = {category: total_matches[category] / total_counts[category] if total_counts[category] > 0 else 0 for category in categories}
recall_total = {category: total_matches[category] / total_counts[category] if total_counts[category] > 0 else 0 for category in categories}
accuracy_total = sum(total_matches.values()) / len(df_filled)

# 创建结果数据框
results_total = pd.DataFrame({
    '类别': categories,
    '精确度（Precision）': [precision_total[category] for category in categories],
    '召回率（Recall）': [recall_total[category] for category in categories],
    'lable列各个类别计数': [total_counts[category] for category in categories]
})

# 打印结果
print(results_total)
print("\n整体准确率（Accuracy）：", accuracy_total)

```
